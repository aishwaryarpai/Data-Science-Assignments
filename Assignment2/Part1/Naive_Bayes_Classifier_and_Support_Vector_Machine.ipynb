```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/content/emails.csv')
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
label=dataset.pop('Prediction')
features=dataset
```

```python
features
```

```python
label
```

```python
features.pop('Email No.')
```

```python
features.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(features,label,test_size=0.2,train_size=0.8,random_state=0)
```

```python
from sklearn.naive_bayes import BernoulliNB
model=BernoulliNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
y_test
```

```python
#Naive Bayes Classifier
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy Score: ',accuracy_score(y_pred,y_test))
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred))
```

```python
#ROC curve for Naive Bayes Classifier
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
print("Area under the curve For Naive Bayers:",roc_auc)
plt.plot(false_positive_rate,true_positive_rate)
```

```python
#Naive Bayes Classifier
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
 from sklearn import svm
```

```python
clf=svm.SVC(kernel='linear')
clf.fit(X_train,y_train)
```

```python
y_pred1=clf.predict(X_test)
```

```python
y_pred1
```

```python
#Support Vector Machine 
print('Accuracy Score: ',accuracy_score(y_pred1,y_test))
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred1))
```

```python
#ROC curve for Support Vector Machine
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred1)
roc_auc = auc(false_positive_rate, true_positive_rate)
print("Area under the curve For Naive Bayers:",roc_auc)
plt.plot(false_positive_rate,true_positive_rate)
```

```python
#Support Vector Machine
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
